##### Importing Matplotlib
Python is very useful because you can load groups of functions created by other people into python and use their code. These groups of programs are called 'modules' in python (think of them as toolboxes that can do different things.) When we downloaded python with Anaconda it contained the most popular modules ('like Matplotlib')

Import statements cam be written as simply as this:
'import module_name'

Sometimes the module was written for a group of different programming languages and kept together. This is the case with matplotlib, so we need to import just the functions related to python. This is why we write the '.pyplot' in the import statement. (Just to tell python only to import the functions for python and not for R, Julia, or other programming languages).

Often, we want to give all of the functions in this module a suffix (also called an alias) to that we know we are using this specific module's functions. In our import statment, we are loading this module with the suffix 'plt'. If we want to use the plot function in matplotlib, we would write 'plt.plot()' to tell python we want it to use the plot function that was loaded from the matplotlib module.

##### Plots are objects
Plots are objects in python. This has two implications:
1. plots have access to many methods that can be used for all plots
    - set_title()
    - set_y_label()
    - set_x_label()
    - set_grid()
    -etc.
2. Plots are named like variables. This makes it very easy to change a plot's properties. For example, you name your plot then you can change it's colors, title or add icons or text to the plot.

#### To make playing with the styles easlier, lets turn the plotting code into a function!

We will create a function that takes as an input three things
 1. The X values
 2. The Y values
 3. The stylesheet name
 
The function will just create the plot. We do not need it to return anything so we will put return None

```python
def plot_function(x, y, style_sheet):
    plt.style.use(style_sheet)
    plt.plot(x, y)
    plt.title('My functions')
    plt.xlabel('This is X')
    plt.ylabel('This is y')
    return None
```

```python
this_stylesheet = 'dark_background'
this_plot = plot_function(x, sin_x, this_stylesheet )
```

```python
this_stylesheet = 'classic'
this_plot = plot_function(x, sin_x, this_stylesheet)
```

```python
this_stylesheet = 'Solarize_Light2'
this_plot = plot_function(x, tan_x, this_stylesheet)
```

## A sighltly more powerful way to plot (Object Oriented Plots)

If we want to put multiple graphs together, then we need to use a <b>Figure</b>. A figure is just the border on an image. Within the figure we can have <b>Axes</b> which are the borders for an individual chart. Within the chart, there are all of the individual pieces of a chart (a y label, a y axis, an x label, and x axis, a title, etc).

#### Details of the object oriented plotting with matplotlib

![Matplotlib diagram](http://matplotlib.org/_images/fig_map.png)

- figure is the outside border of the chart or charts. This is the area that gets printed or saved so it includes everything inside it's borders. A figure can contain zero to very many charts. Any charts contained within the figure are considered 'children' of the figure. (And the figure is there 'parent')

To create a figure the command is plt.figure()

- axes are the individual chart or each chart if there are multiple charts

- axis are the x-axis and y-axis of an individual axes

###### plt.subplots will create a figure and axes based on the parameters input into subplots.

The first parameter is an integer representing how many rows of charts you would like, the second parameter is an integer representing how many columns of charts you would like.

It is very helpful to add a tuple called figsize - which will control the figure size. The first element of the tuple controls the width of the figure, the second controls the height of the figure

```python
# To create a figure and axes, the plt.subplots function is most oftenly used
fig, axs = plt.subplots(1,1, figsize=(12,5))
```

```python
fig, axs = plt.subplots(1,1, figsize=(10,4))
axs = plt.plot(x, tan_x)
```

### Multiple charts
You can change the figure to contain multiple plots by chaning the integers you feed into plt.subplots

The axes variable will be indexed just like a numpy array. The code below plots our geometric functions. Notice how they are referenced by calling axes[index].plot() 

```python
fig, axes = plt.subplots(1,3, figsize = (12,4))
axes[0].plot(x, sin_x)
axes[1].plot(x, cos_x)
axes[2].plot(x, tan_x)
```

### Cusomizing the features of the charts contains a slightly different syntax when done using object oriented plots

The commands for customizing the charts use 'set_' before the attribute you would like to set.
Below we will be adding titles, x-axis labels, y_axis labels,

```python
fig, axes = plt.subplots(1,3, figsize = (12,4))
axes[0].plot(x, sin_x)
axes[0].set_title('This is sin')
axes[0].set_xlabel('x_axis in units of pi')
axes[0].set_ylabel('sin(x)')

axes[1].plot(x, cos_x)
axes[1].set_title('This is cos')
axes[1].set_xlabel('x_axis in units of pi')
axes[1].set_ylabel('cos(x)')

axes[2].plot(x, tan_x)
axes[2].set_title('This is tan')
axes[2].set_xlabel('x_axis in units of pi')
axes[2].set_ylabel('tan(x)')
```

### So how does it work when there are multiple columns and rows?
(Hint: Just like numpy arrays)

```python
# note: the standard notation you will see abbreviates axes with axs. 
# When you are googling stack overflow or the matplotlib library, it will look like this 99% of the time
# So we will also be using this notation
fig, axs = plt.subplots(2,2, figsize=(12,6))

# Assign upper left chart to be sin(x)
axs[0,0].plot(x, sin_x)
axs[0,0].set_title('This is sin')
axs[0,0].set_xlabel('x_axis in units of pi')
axs[0,0].set_ylabel('sin(x)')

# assign lower left chart to be cos_x
axs[1,0].plot(x, cos_x)
axs[1,0].set_title('This is cos')
axs[1,0].set_xlabel('x_axis in units of pi')
axs[1,0].set_ylabel('cos(x)')

# assign lower right chart to be tan_x
axs[1,1].plot(x, tan_x)
axs[1,1].set_title('This is tan')
axs[1,1].set_xlabel('x_axis in units of pi')
axs[1,1].set_ylabel('tan(x)')

#plt.tight_layout()
```

#### Bar Charts
Plot a single variable as a bar chart is easy. There is a slight nuance to plotting stacked bar charts or multiple bars on a chart.

```python
# Creating x and y (random) variables to plot
x_bar = np.arange(1, 21)
y_bar1 = np.random.randint(0,40, 20)
y_bar2 = np.random.randint(0,30, 20)
y_bar3 = np.random.randint(0,10, 20)
```

```python
print(x_bar)
```

##### Single bar chart

```python
fig, axs = plt.subplots(1,1, figsize=(10,5))
axs.bar(x_bar, y_bar)
axs.set_xlim(0, 21)
axs.set_title('Simple Single Bar Chart')
```

##### Stacked bar chart

```python
fig, axs = plt.subplots(1,1, figsize=(10,5))
axs.bar(x_bar, y_bar)
axs.bar(x_bar, y_bar2)
axs.bar(x_bar, y_bar3)
axs.set_xlim(0, 21)
axs.set_title('Stacked Bar Chart')
```

### To create a grouped bar chart, you need to do a little more work specifying where to draw the bars for each series being plotted.

We add the variable width which determines how wide each bar column is AND we also use this to shift the x-axis for each series such that they do not overlap.

```python
width = .25

fig, axs = plt.subplots(1,1, figsize=(10,5))

axs.bar(x_bar-width, y_bar, width)
axs.bar(x_bar, y_bar2, width)
axs.bar(x_bar+width, y_bar3, width)
axs.set_xlim(0, 21)
axs.set_title('Grouped Bar Chart')
```

### Using Numpy and Boolean Masks to add more depth/information to your plots

Numpy boolean masks can be created to change the plotted colors.
For example, suppose we are interested in highlighting any y-values that are above a threshold.

First create a boolean mask for these items

```python
y_above_25 = y_scatter>25
y_above_25
```

```python
y_below_zero = y_scatter < 0
y_below_zero
```

Now we are going to use a list of colors based on these masks.

```python
color_list = []
for i in range(len(x_scatter)):
    if y_below_zero[i]==True:
        color_list.append('red')
    elif y_above_25[i]==True:
        color_list.append('green')
    else:
        color_list.append('blue')
color_list
```

```python
fig, axs = plt.subplots(1,1, figsize = (10,5))
axs.scatter(x_scatter, y_scatter, c=color_list)
axs.set_xlim(0, 21)
axs.set_title('Simple Scatter Plot')
```

##### Below is a simple way to iterate over the axis and create multiple plots.

Create x_data which is about -pi to pi

```python
x_data = np.linspace(-3.14, 3.14, 100)
x_data
```

Create a list of functions you want to plot

```python
y_funcs = [np.sin, np.cos, np.sqrt, np.degrees]
```

```python
fig, axs = plt.subplots(2, 2)

for ax, y_func in zip(axs.flatten(), y_funcs):
    ax.scatter(x_data, y_func(x_data))
      
fig.suptitle('Our four functions')
```
